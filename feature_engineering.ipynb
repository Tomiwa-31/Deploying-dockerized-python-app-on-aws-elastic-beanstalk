Creating temporal,behavioural and transformational feature

```python
def create_temporal_feature(df):
    df = df.copy()

    # Convert to DateTime
    df['date'] = pd.to_datetime(df['date'])

    # Basic Time Features
    df['hour'] = df['date'].dt.hour
    df['day_of_week'] = df['date'].dt.dayofweek
    df['day_of_month'] = df['date'].dt.day
    df['month'] = df['date'].dt.month

    # Fraud specific time Features
    df['is_weekend'] = (df['day_of_week'] >= 5).astype('int8')
    df['is_night'] = ((df['hour'] >= 0) & (df['hour'] <= 5)).astype('int8')
    df['is_business_hours'] = ((df['hour'] >= 9) & (df['hour'] <= 17)).astype('int8')

    # Sort feature for client id and date
    df = df.sort_values(['client_id','date'])

    # MEMORY-EFFICIENT: Process in chunks for rolling features
    df = add_rolling_features_memory_efficient(df)

    # Outlier Detection
    df['amount_zscore'] = (df['amount'] - df['amount'].mean()) / df['amount'].std()
    df['is_amount_outlier'] = (df['amount_zscore'].abs() > 3).astype('int8')

    return df
def add_rolling_features_memory_efficient(df, chunk_size=10000):
    """Add rolling features with smaller chunks"""

    results = []
    print(f"Processing {len(df)} rows in chunks of {chunk_size}...")

    for i in range(0, len(df), chunk_size):
        chunk = df.iloc[i:i + chunk_size].copy()
        chunk = chunk.sort_values(['client_id', 'date'])

        # Set index for rolling
        chunk_indexed = chunk.set_index('date')

        try:
            # Rolling features with smaller chunks
            chunk['recent_tx_count_1d'] = chunk_indexed.groupby('client_id')['amount'].rolling('1D').count().values
            chunk['recent_avg_amount_7d'] = chunk_indexed.groupby('client_id')['amount'].rolling('7D').mean().values
        except MemoryError:

            chunk['recent_tx_count_1d'] = chunk.groupby(['client_id', chunk['date'].dt.date])['id'].transform('count')
            chunk['recent_avg_amount_7d'] = chunk.groupby('client_id')['amount'].transform('mean')

        results.append(chunk)

        if i % 100000 == 0:
            print(f"Processed rows {i} to {min(i + chunk_size, len(df))}")

    return pd.concat(results, ignore_index=True)

```

```python
merge_df_new_features=create_temporal_feature(merge_df)
```
